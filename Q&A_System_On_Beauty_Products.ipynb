```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

```python
from nltk.tokenize import word_tokenize
def tokenize(text):
    return word_tokenize(text)
```

```python
def toLowercase(tokenized_words):
    return [tokens.lower() for tokens in tokenized_words]
```

```python
import re
from string import punctuation
def punctuation(lowercase_tokens):
    return [re.sub(r'([^\s\w]|_)+', '', tokens) for tokens in lowercase_tokens]
```

```python
from nltk.corpus import stopwords
def remove_stopwords(tokenized_words):
    stop_words = stopwords.words('english')
    return [tokens for tokens in tokenized_words if not tokens in stop_words]
```

```python
from spellchecker import SpellChecker
def spell_checker(trimmed_tokens):
    spell = SpellChecker()
    return [spell.correction(tokens) for tokens in trimmed_tokens]
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
def get_wordnet_pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)

def lemmatize(correct_words):
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = ' '.join([lemmatizer.lemmatize(word,get_wordnet_pos(word)) for word in correct_words])
    return lemmatized_tokens
```

```python
def find_WH(list_words):
    WH_postag = ['WDT', 'WP', 'WP$', 'WRB']
    postag = nltk.pos_tag(list_words)
    WH_Question = [True for value in postag if value[1] in WH_postag]
    return WH_Question
```

```python
def preprocessing(data):
    tokenized_words=tokenize(data)
    lowercase_tokens=toLowercase(tokenized_words)
    noPunch_tokens=punctuation(lowercase_tokens)
    correct_words=spell_checker(noPunch_tokens)
    trimmed_tokens=remove_stopwords(correct_words)
    lemmatized_words=lemmatize(trimmed_tokens)
    #PP_data = [word for word in lemmatized_tokens if word]
    WH_flag=find_WH(lemmatized_words)

    return trimmed_tokens,WH_flag,lemmatized_words
```

```python
from textblob import TextBlob
def sentiment_analyser(dataframe, flag):
    
    if flag == 'q':
        q_sentiment_score = dataframe['answer'].apply(lambda x: TextBlob(x).sentiment[0] )
        print("q_Sentiment_Score",q_sentiment_score)
        qa_sentiments = ['positive sentiment' if score >= 0 else 'negative sentiment' for score in q_sentiment_score]
        return qa_sentiments
    elif flag == 'r':
        r_sentiment_score = dataframe['reviewText'].apply(lambda x: TextBlob(x).sentiment[0] )
        print("r_Sentiment_Score",r_sentiment_score)
        r_sentiments = ['positive sentiment' if score >= 0 else 'negative sentiment' for score in r_sentiment_score]
       # list3 = []
       # r_sentiment_score=r_sentiment_score.values.tolist()
        #temp = [1 if i >= 0 else 0 for i in r_sentiment_score]
        #for i in range(len(dataframe)):
         #   list3.append(dataframe['overall'][i] + r_sentiment_score[i])
          #  review_sentiments = ['positive sentiment' if value >= 3 else 'negative sentiment' for value in list3]
        return r_sentiments
```

```python
def question_preprocess(QA_filtered_df):
    tokenised_question = []
    preprocess_question=[]
    question_type = []
    for questions in QA_filtered_df:
        preprocess_question=preprocessing(questions)
        tokenised_question.append(preprocess_question[2])
        WH_flag = preprocess_question[1]
        if len(WH_flag) == 0:
            question_type.append("yes/no")
        else:
            question_type.append("open-ended")
    return tokenised_question,question_type


def review_preprocess(review_filtered_df):
    tokenised_review = []
    for review in review_filtered_df:
        preprocess_review=preprocessing(review)
        tokenised_review.append(preprocess_review[2])
    return tokenised_review


QA_df = QA_dataset.copy()
#QA_df=QA_df[:][:15]
review_df = review_dataset.copy()
#review_df=review_df[:][:15]
tokenised_question = question_preprocess(QA_df['question'])
QA_df['tokenised question'] = tokenised_question[0]
print("Tokenized question 0 ",tokenised_question[0])
print("Tokenized question 1 ",tokenised_question[1])

tokenised_review = review_preprocess(review_df['reviewText'])
review_df['tokenised reviews'] = tokenised_review
print("Tokenized review : ",tokenised_review)

QA_df['question_type'] = QA_df['questionType']
```

```python
QA_df['answer_sentiments'] = sentiment_analyser(QA_df, 'q')
review_df['review_sentiments'] = sentiment_analyser(review_df, 'r')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
def tf_idf(query,data):
    tf_idf = TfidfVectorizer(analyzer='word')
    tfidf_matrix =  tf_idf.fit_transform(data)
    score_tf_idf=tf_idf.transform([query])
    feature_names = tf_idf.get_feature_names()
    for col in score_tf_idf.nonzero()[1]:
        score_tf=feature_names[col], ' - ', score_tf_idf[0, col]
    return score_tf
    
```

```python
from rank_bm25 import BM25Okapi
def bm25(query,data):
    corpus = data
    tokenized_query = query.split(" ")
    tokenized_corpus = [doc.split(" ") for doc in corpus]

    bm25 = BM25Okapi(tokenized_corpus)
        
    bm25_scores = bm25.get_scores(tokenized_query)
    #print("BM25 score: ",bm25_scores)

    #relevant_query=bm25.get_top_n(tokenized_query, corpus, n=3)
    #print(relevant_query)
       # df_new['relevance_score'] = bm25_scores
        
       # df_new.sort_values(['relevance_score'],ascending=False, inplace=True)
        #print("New datafile :",df_new)
    return bm25_scores
```

```python
def relevance_check(query,dataframe,flag):
    df_new=dataframe.copy()
    if flag=='q':
        data=df_new['answer']
        tf_idf_score=tf_idf(query,data)
        data_relevance_score=bm25(query,data)
        df_new['relevance_score'] =  data_relevance_score
        df_new.sort_values(['relevance_score'],ascending=False, inplace=True)
        
    else:
        data=df_new['reviewText']
        tf_idf_score=tf_idf(query,data)
        data_relevance_score=bm25(query,data)
        df_new['relevance_score'] =  data_relevance_score
        df_new.sort_values(['relevance_score'],ascending=False, inplace=True)
    return df_new
```

```python
def review_generator(dataframe, answerSentiment):
    review1 = dataframe['reviewText'].iloc[0]
    review1Sentiments = dataframe['review_sentiments'].iloc[0]
    review1RelScore = dataframe['relevance_score'].iloc[0]
    if answerSentiment == 'positive sentiment':
        ndf = dataframe[dataframe['review_sentiments'] == 'negative sentiment']
        if len(ndf) == 0:
            review2 = dataframe['reviewText'].iloc[1]
            review3 = dataframe['reviewText'].iloc[2]
            review2Sentiments = dataframe['review_sentiments'].iloc[1]
            review3Sentiments = dataframe['review_sentiments'].iloc[2]
            review2RelScore = dataframe['relevance_score'].iloc[1]
            review3RelScore = dataframe['relevance_score'].iloc[2]
        elif len(ndf) == 1:  # Only 1 negative review
            review2 = dataframe['reviewText'].iloc[1]
            review3 = ndf['reviewText'].iloc[0]
            review2Sentiments = dataframe['review_sentiments'].iloc[1]
            review3Sentiments = ndf['review_sentiments'].iloc[0]
            review2RelScore = dataframe['relevance_score'].iloc[1]
            review3RelScore = ndf['relevance_score'].iloc[0]
        else:  # More than 1 negative review
            review2 = ndf['reviewText'].iloc[0]
            review3 = ndf['reviewText'].iloc[1]
            review2Sentiments = ndf['review_sentiments'].iloc[0]
            review3Sentiments = ndf['review_sentiments'].iloc[1]
            review2RelScore = ndf['relevance_score'].iloc[0]
            review3RelScore = ndf['relevance_score'].iloc[1]

    elif answerSentiment == 'negative sentiment':
        ndf = dataframe[dataframe['review_sentiments'] == 'positive sentiment']

        if len(ndf) == 0:  # No positive reviews
            review2 = dataframe['reviewText'].iloc[1]
            review3 = dataframe['reviewText'].iloc[2]
            review2Sentiments = dataframe['review_sentiments'].iloc[1]
            review3Sentiments = dataframe['review_sentiments'].iloc[2]
            review2RelScore = dataframe['relevance_score'].iloc[1]
            review3RelScore = dataframe['relevance_score'].iloc[2]
        elif len(ndf) == 1:  # Only 1 positive review
            review2 = dataframe['reviewText'].iloc[1]
            review3 = ndf['reviewText'].iloc[0]
            review2Sentiments = dataframe['review_sentiments'].iloc[1]
            review3Sentiments = ndf['review_sentiments'].iloc[0]
            review2RelScore = dataframe['relevance_score'].iloc[1]
            review3RelScore = ndf['relevance_score'].iloc[0]
        else:  # More than 1 positive review
            review2 = ndf['reviewText'].iloc[0]
            review3 = ndf['reviewText'].iloc[1]
            review2Sentiments = ndf['review_sentiments'].iloc[0]
            review3Sentiments = ndf['review_sentiments'].iloc[1]
            review2RelScore = ndf['relevance_score'].iloc[0]
            review3RelScore = ndf['relevance_score'].iloc[1]

    elif answerSentiment == 'None':
        review2 = dataframe['reviewText'].iloc[1]
        review3 = dataframe['reviewText'].iloc[2]
        review2Sentiments = dataframe['review_sentiments'].iloc[1]
        review3Sentiments = dataframe['review_sentiments'].iloc[2]
        review2RelScore = dataframe['relevance_score'].iloc[1]
        review3RelScore = dataframe['relevance_score'].iloc[2]

    reviews = [review1, review2, review3]
    revSentiments = [review1Sentiments, review2Sentiments, review3Sentiments]
    revRelScore = [review1RelScore, review2RelScore, review3RelScore]

    return reviews, revSentiments, revRelScore
```

```python
def present_in_both(processed_query, WH_flag):
    if len(WH_flag) == 0:
        newDF = QA_df[QA_df['question_type'] == "yes/no"]
    else:
        newDF = QA_df[QA_df['question_type'] == "open-ended"]

    if len(newDF) == 0:
        answerJson = {'answer': "Sorry! No relevant answer found for your query.!!",'answer_sentiments': None, 'qaRelScore': None}
    else:
        flag = 'q'
        df = relevance_check(processed_query, newDF, flag)
        answer = df['answer'].iloc[0]
        ansSentiment = df['answer_sentiments'].iloc[0]
        ansRelScore = df['relevance_score'].iloc[0]
        
        if ansRelScore == 0:
            answerJson = {'answer': "Sorry! No relevant answer found for your query.!!",'answer_sentiments': None, 'qaRelScore': None}
        else:
            answerJson = {'answer': answer, 'answer_sentiments': ansSentiment,'qaRelScore': ansRelScore}

    flag = 'r'
    df = relevance_check(processed_query, RV_df, flag)
    reviews, revSentiments, revRelScore = review_generator(df, ansSentiment)
        
    if revRelScore[0] == 0:
        msg = ["Sorry! No relevant reviews found!"]
        reviewJson = {'reviews': msg,'review_sentiments': None, 'rvRelScore': None}
    else:
        reviewJson = {'reviews': reviews,'review_sentiments': revSentiments, 'rvRelScore': revRelScore}

    return answerJson, reviewJson
```

```python
def present_only_in_rv(processed_query):
    answerJson = {'answer': None,'answer_sentiments': None, 'qaRelScore': None}
    flag = 'r'
    ansSentiment = 'None'
    df = relevance_check(processed_query, RV_df, flag)
    reviews, revSentiments, revRelScore = review_generator(df, ansSentiment)
    if revRelScore[0] == 0:
        msg = ["Sorry! No relevant reviews found!"]
        reviewJson = {'reviews': msg,'review_sentiments': None, 'rvRelScore': None}
    else:
        reviewJson = {'reviews': reviews,'review_sentiments': revSentiments, 'rvRelScore': revRelScore}

    return answerJson, reviewJson
```

```python
def present_only_in_qa(processed_query, WH_flag):
    if len(WH_flag) == 0:
        newDF = QA_df[QA_df['question_type'] == "yes/no"]
    else:
        newDF = QA_df[QA_df['question_type'] == "open-ended"]

    if len(newDF) == 0:
        answerJson = {'answer': "Oops! No relevant answer found for your question type!",'answer_sentiments': None, 'qaRelScore': None}
    else:
        flag = 'q'
        df = relevance_check(processed_query, newDF, flag)
        answer = df['answer'].iloc[0]
        ansSentiment = df['answer_sentiments'].iloc[0]
        ansRelScore = df['relevance_score'].iloc[0]

        if ansRelScore == 0:
            answerJson = {'answer': "Sorry! No relevant answer found for your question.",'answer_sentiments': None, 'qaRelScore': None}
        else:
            answerJson = {'answer': answer, 'answer_sentiments': ansSentiment,'qaRelScore': ansRelScore}

    reviewJson = {'reviews': None,'review_sentiments': None, 'rvRelScore': None}

    return answerJson, reviewJson

   
```

```python
 def not_present_in_both():
    answerJson = {'answer': None,'answer_sentiments': None, 'qaRelScore': None}
    reviewJson = {'reviews': None,'review_sentiments': None, 'rvRelScore': None}

    return answerJson, reviewJson
```

```python
def find_relevant_stuff(processed_query, WH_flag):
    val = check_asin_id()
    if val == 0:
        relAnswer, relReviews = present_in_both(processed_query, WH_flag)
    elif val == 1:
        relAnswer, relReviews = present_only_in_rv(processed_query)
    elif val == 2:
        relAnswer, relReviews = present_only_in_qa(processed_query, WH_flag)
    elif val == 3:
        relAnswer, relReviews = not_present_in_both()
    
    return relAnswer, relReviews
```

```python
def merged(dic1, dic2):
    dic = dic1.copy()
    dic.update(dic2)
    return dic
```

```python
def query_processor(query):
    processed_query, WH_flag,lemmatized_word = preprocessing(query)
    if len(query) == 0:
        answerJson = {'answer': None,'answer_sentiments': None, 'qaRelScore': None}
        reviewJson = {'reviews': None,'review_sentiments': None, 'rvRelScore': None}
        sendBackJson = merged(answerJson, reviewJson)
        return sendBackJson
    relAnswer, relReviews =find_relevant_stuff(query,WH_flag)
    sendBackJson = merged(relAnswer, relReviews)
    return sendBackJson
```
